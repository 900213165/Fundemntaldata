```python
df=pd.read_csv('covid_data.csv',encoding="latin-1")
```

```python
df.head()
```

```python
from pandas.api.types import CategoricalDtype
cats=["Fri", "Sat", "Sun", "Mon", "Tue", "Wed","Thus"]
cat_type=CategoricalDtype(categories=cats, ordered=True)
df["weekdays"]=df["weekdays"].astype(cat_type)
```

```python
dfegy=df[df["country"]=="Egypt"]
```

```python
stats=dfegy.groupby("weekdays").agg({"dcases":[np.mean, np.std, np.size]})
```

```python
stats
```

```python
import numpy as np
import scipy 
```

```python
stats.index
```

```python
ci95_hi = []
ci95_lo = []
```

```python
import statsmodels.api as sm
from statsmodels.stats.proportion import proportion_confint
```

```python
for i in stats.index:
    m, s, n = stats.loc[i]
    x=scipy.stats.t.interval(.95, n-1, m,s/np.sqrt(n-1))
    ci95_hi.append(x[1])
    ci95_lo.append(x[0])
```

```python
ci95_hi
```

```python
ci95_lo
```

```python
stats['ci95_hi'] = ci95_hi
stats['ci95_lo'] = ci95_lo
print(stats)
```

```python
df_ci= pd.DataFrame(stats)
```
